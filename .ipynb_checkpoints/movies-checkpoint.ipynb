```python
import pandas as pd
import numpy as np
```

## Import data

```python
movies = pd.read_csv("ml-25m/movies.csv")
movies.head()
```

## Data Wrangling

```python
# replace missing genre with np.nan
movies["genres"] = movies["genres"].replace("(no genres listed)", np.nan)
movies
```

```python
# count amount of nan
movies.isna().sum()
```

```python
# remove all nan rows
movies = movies[movies["genres"].notna()]
movies
```

```python
# check year
year_lst = []

for movie in movies["title"]:
    if movie[-1] == ")":
        year = movie[-6:]
        year = year[1:-1]
    else:
        year = np.nan
        
    year_lst.append(year)
```

```python
movie_lst = []
year_lst = []

for movie in movies["title"]:
    # remove year from movie title
    movie_title = movie[:-6]
    movie_lst.append(movie_title)
    
    # remove movie title from year
    year = movie[-6:]
    # remove paranthesis
    year = year[1:-1]
    
    year_lst.append(year)
```

```python
#movies["title"] = movie_lst
movies["year"] = year_lst
movies
```

```python
# reorder columns
movies = movies[["title", "year", "genres"]]
movies
```
