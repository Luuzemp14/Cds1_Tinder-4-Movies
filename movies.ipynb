```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("ml-25m/movies.csv")
movies.head()
```

```python
# set movie id as index
movies = movies.set_index("movieId")
movies.head()
```

```python
movie_lst = []
year_lst = []

for movie in movies["title"]:
    # remove year from movie title
    movie_title = movie[:-6]
    movie_lst.append(movie_title)
    
    # remove movie title from year
    year = movie[-6:]
    # remove paranthesis
    year = year[1:-1]
    
    year_lst.append(year)
```

```python
movies["title"] = movie_lst
movies["year"] = year_lst
movies
```

```python
# reorder columns
movies = movies[["title", "year", "genres"]]
movies
```
