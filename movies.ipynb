```python
import pandas as pd
import numpy as np
import re
```

## Import data

```python
movies = pd.read_csv("ml-25m/movies.csv")
movies.head()
```

## Data Wrangling

```python
# replace missing genre with np.nan
movies["genres"] = movies["genres"].replace("(no genres listed)", np.nan)
movies
```

```python
# count amount of nan
movies.isna().sum()
```

```python
# extract year
movies['year'] = movies['title'].str.extract(r"\((\d{4})\)", expand = False)
movies
```

```python
# remove all nan rows
movies = movies[movies["genres"].notna()]
movies = movies[movies["year"].notna()]
movies
```

```python
# year to int
movies["year"] = movies["year"].astype("int")
```

```python
movies["year"].dtype
```

```python
# reorder columns
movies = movies[["title", "year", "genres"]]
movies
```

```python
print(movies["year"].unique())
```
