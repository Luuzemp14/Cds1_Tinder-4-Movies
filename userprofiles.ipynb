```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Movie data

```python
movies = pd.read_csv("ml-25m/movies.csv")
movies.head()
```

### 1.1 Data Wrangling

```python
# replace missing data in genre column with NAN
movies["genres"] = movies["genres"].replace("(no genres listed)", np.nan)
```

```python
# count amount of nan
movies.isna().sum()
```

```python
# extract year
movies['year'] = movies['title'].str.extract(r"\((\d{4})\)")
movies
```

```python
# besser wäre es, die jahren im titel zu finden und dann direkt eine neue spalte zu erstellen 
# UND aus dem titel entfernen.
```

```python
# remove all nan rows
movies = movies[movies["genres"].notna()]
movies = movies[movies["year"].notna()]

# year to int
movies["year"] = movies["year"].astype("int")

# reorder columns
movies = movies[["movieId", "title", "year", "genres"]]
```

```python
# filter movies released between 2010 and 2020
movies_2010 = movies.loc[(movies["year"] >= 2010) & (movies["year"] <= 2020)]
movies_2010
```
