## 1. Movie data
Wir importieren Movielens Daten und verwenden nur die Movies, die in den abgefragten tmdb Daten vorhanden sind.

```python
df_movies = pd.read_csv("ml-25m/movies.csv")
df_links = pd.read_csv("ml-25m/links.csv")
```

```python
# merge both dataframes on movieId column
df_movies = pd.merge(df_movies, df_links, on='movieId', how='outer')
```

```python
df_movies = df_movies[["movieId", "tmdbId", "title", "genres"]]
df_movies
```

```python
tmdb_movies = pd.read_csv("tmdb_movies.csv", sep="\t")

# rename id column so that we can merge later
tmdb_movies.rename(columns={'id':'tmdbId'}, inplace=True)
```

```python
# merge movielens movies with tmdb movies
df_movies2 = pd.merge(df_movies, tmdb_movies, on='tmdbId')
```

```python
df_movies2
```

```python
## 13 zu viele Zeilen --> Duplikate!! Aber warum??
```

```python
# extract data from dictionaries and list, separate key values by '|'

def extract_values(data):
    return data.apply(lambda x: '|'.join([i['name'] for i in eval(x)]))

df_movies2['genres_y'] = extract_values(df_movies2["genres_y"])
df_movies2['spoken_languages'] = extract_values(df_movies2["spoken_languages"])
df_movies2['production_companies'] = extract_values(df_movies2['production_companies'])
df_movies2['production_countries'] = extract_values(df_movies2['production_countries'])

# belongs_to_collection
```

```python
# adult is dropped because there are only 3 adult films. which makes it useless to use for our model
cols = [
    "title_x",
    "genres_x",
    "Unnamed: 0",
    "backdrop_path",
    "homepage",
    "poster_path",
    "imdb_id",
    "original_title",
    "video",
    "status",
    "adult"
]

# drop unused columns
df = df_movies2.drop(df_movies2[cols], axis=1)

# rename columns
df.rename(columns={'genres_y':'genres', 'title_y':'title'}, inplace=True)

# reorder columns
col_order = [
    "movieId", 
    "title", 
    "genres", 
    "overview", 
    "tagline",
    "release_date", 
    "vote_average", 
    "vote_count", 
    "popularity", 
    "runtime", 
    "original_language",
    "spoken_languages", 
    "production_companies",
    "production_countries",
    "revenue",
    "budget",
    "belongs_to_collection",
    
]

df2 = df[col_order]
df2
```

```python
# replace empty field with np.nan
df2 = df2.replace("", np.nan)
```

## Drop Rows

```python
# drop duplicates
df2 = df2.drop_duplicates('movieId')

# drop rows with NAN 
df2.dropna(subset=['overview', 'genres'])
```
